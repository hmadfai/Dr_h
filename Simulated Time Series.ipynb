```python
# prompt: simulate a time series from SARIMA(.2, 1, .3)(-.4,2,.4) with seasonality of 7 days. Plot the simulated data

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Set parameters for SARIMA model
ar_params = [0.2]
ma_params = [0.3]
seasonal_ar_params = [-0.4]
seasonal_ma_params = [0.4]
order = (1, 1, 1)
seasonal_order = (1, 2, 1, 7)  # (P, D, Q, s) where s is the seasonal period (7 days)

# Generate random noise
noise = np.random.normal(loc=0, scale=1, size=100)

# Simulate the time series using SARIMAX
model = SARIMAX(noise, order=order, seasonal_order=seasonal_order)
results = model.fit(disp=False)
simulated_data = results.simulate(nsimulations=100)

# Plot the simulated data
plt.plot(simulated_data)
plt.title('Simulated Time Series from SARIMA(.2, 1, .3)(-.4,2,.4) with 7-day Seasonality')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
